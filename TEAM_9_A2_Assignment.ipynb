## Merge Columns from Datasets

```python
# Adding the Restaurant Name

df_res = (pd.merge(df_orders,
                   df_restaurants[['RESTAURANT_ID','NAME']],
                   left_on='RESTAURANT_ID',
                   right_on='RESTAURANT_ID',
                   how='left')
            .rename(columns={'NAME':'RESTAURANT_NAME'}))

# Display the result
df_res.head()
```

```python
# Adding the name of the MAIN

df_res1 = (pd.merge(df_res,
          df_items[['ITEM_ID', 'ITEM_NAME']],
          left_on='MAIN',
          right_on='ITEM_ID',
          how='left')
.rename(columns={'ITEM_NAME':'MAIN_NAME'})
.drop(columns='ITEM_ID'))

# Display the result
df_res1.head()
```

```python
# Dropping ITEM_ID

df_res2 = (pd.merge(df_res1,
          df_items[['ITEM_ID', 'ITEM_NAME']],
          left_on='BASE',
          right_on='ITEM_ID',
          how='left')
.rename(columns={'ITEM_NAME':'BASE_NAME'})
.drop(columns='ITEM_ID'))

# Display the result
df_res2.head()
```

## Drop " N/A" and "0" Values



```python
# Drop rows where DRINKS and COOKIES are 0.0, and MAIN, BASE, SIDE_1, and SIDE_2 are all NaN

df_res3 = df_res2[
    ~((df_res2['DRINKS'] == 0.0) &
      (df_res2['COOKIES'] == 0.0) &
      df_res2[['MAIN', 'BASE', 'SIDE_1', 'SIDE_2']].isna().all(axis=1))
]

# Display the result
df_res3
```

```python
# Average number of orders per day for each restaurant

# Ensure 'DATETIME' column is in datetime format
df_orders['DATETIME'] = pd.to_datetime(df_orders['DATETIME'])

# Extract the day of the week
df_orders['DAY_OF_WEEK'] = df_orders['DATETIME'].dt.day_name()

# Group by 'RESTAURANT_ID' and 'DAY_OF_WEEK', then count the number of orders
orders_per_day_per_restaurant = df_orders.groupby(['RESTAURANT_ID', 'DAY_OF_WEEK'])['ORDER_ID'].count().reset_index(name='ORDER_COUNT')

# Calculate the average orders per day for each restaurant
average_orders_per_day_per_restaurant = orders_per_day_per_restaurant.groupby(['RESTAURANT_ID', 'DAY_OF_WEEK']).mean().reset_index()

# Group by 'RESTAURANT_ID' and 'DAY_OF_WEEK' and count the number of orders
orders_per_day_per_restaurant = df_orders.groupby(['RESTAURANT_ID', 'DAY_OF_WEEK'])['ORDER_ID'].count().reset_index(name='ORDER_COUNT')

# Pivot the data to make 'DAY_OF_WEEK' columns and 'RESTAURANT_ID' rows
average_orders_pivot = orders_per_day_per_restaurant.pivot(index='RESTAURANT_ID', columns='DAY_OF_WEEK', values='ORDER_COUNT')

# Fill any NaN values with 0 (in case any restaurant didn't have orders on a specific day)
average_orders_pivot = average_orders_pivot.fillna(0)

# Display the result
print(average_orders_pivot)
```

```python
# Average number of order per each opening hour for each day across the week (Visualization in next section)

# Convert 'DATETIME' column to datetime format
df_orders['DATETIME'] = pd.to_datetime(df_orders['DATETIME'])

# Extract day of the week (e.g., Monday, Tuesday)
df_orders['DAY_OF_WEEK'] = df_orders['DATETIME'].dt.day_name()

# Extract hour of the day (0-23)
df_orders['HOUR'] = df_orders['DATETIME'].dt.hour

# Group by 'DAY_OF_WEEK' and 'HOUR' and count the orders
orders_per_hour_per_day = df_orders.groupby(['DAY_OF_WEEK', 'HOUR']).size().reset_index(name='ORDER_COUNT')

# Calculate the average number of orders per hour per day
average_orders = orders_per_hour_per_day.groupby(['DAY_OF_WEEK', 'HOUR'])['ORDER_COUNT'].mean().reset_index()

# Pivot the table
pivot_table = average_orders.pivot(index='HOUR', columns='DAY_OF_WEEK', values='ORDER_COUNT')

# Reorder the days of the week if necessary
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Display the Result
pivot_table = pivot_table[day_order]
print(pivot_table)
```

```python
# Highlighting Least busy and busiest hour of each day for all restaurants

# Convert 'DATETIME' column to datetime format
df_orders['DATETIME'] = pd.to_datetime(df_orders['DATETIME'])

# Extract day of the week (e.g., Monday, Tuesday)
df_orders['DAY_OF_WEEK'] = df_orders['DATETIME'].dt.day_name()

# Extract hour of the day (0-23)
df_orders['HOUR'] = df_orders['DATETIME'].dt.hour

# Group by 'DAY_OF_WEEK' and 'HOUR' and count the orders
orders_per_hour_per_day = df_orders.groupby(['DAY_OF_WEEK', 'HOUR']).size().reset_index(name='ORDER_COUNT')

# Calculate the average number of orders per hour per day
average_orders = orders_per_hour_per_day.groupby(['DAY_OF_WEEK', 'HOUR'])['ORDER_COUNT'].mean().reset_index()

# Pivot the table
pivot_table = average_orders.pivot(index='HOUR', columns='DAY_OF_WEEK', values='ORDER_COUNT')

# Reorder the days of the week if necessary
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
pivot_table = pivot_table[day_order]

# Function to highlight min and max values in yellow for each day
def highlight_min_max(s):
    is_min = s == s.min()
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in (is_min | is_max)]

# Apply the highlighting function to the pivot table
styled_table = pivot_table.style.apply(highlight_min_max, axis=0)

# Display the styled table
styled_table
```

It is evident that across the week, opening hour is the least busy by a landslide, while noon launches the busiest hour of the day.
